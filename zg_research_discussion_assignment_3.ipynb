**Topic:** Bias and Ethics in Recommender Systems

I came into this week thinking bias in recommender systems was mostly a data problem — bad training data in, bad recommendations out. SVD's energy rule complicates that. The whole point of SVD is to find the latent structure that explains the most variance in the data. We keep the top *k* singular values — the ones that satisfy the 90% energy rule from the readings — and throw the rest away. But variance isn't neutral. It's dominated by majority patterns.

If the training data reflects a world where men were historically steered toward tech jobs, that signal shows up across thousands of users. It's high-energy — the algorithm keeps it. The Carnegie Mellon AdFisher study, cited in Estola's talk, documented this directly — identical fake profiles differing only in specified gender were shown far fewer high-paying executive job ads by Google's system. The preference signals of underrepresented groups are, by contrast, sparse and low-variance. They get compressed out when you truncate to *k* dimensions.

What makes this harder than it looks is that it's not just biased input data — the optimization objective itself treats majority patterns as signal. SGD and ALS both minimize aggregate RMSE, which means they're most accurate for users with the most ratings and least accurate for underrepresented groups. The Hardt et al. paper proposes equalized odds as a post-hoc fix, but in a latent factor space you can't just retrain on cleaner data — the constraint would have to be built into the decomposition itself, which changes the whole objective.

I keep wondering if Estola's ensemble approach from the video is doing something structurally similar: not letting any single latent structure dominate. Does that actually help, or does it just average out multiple biased models?
