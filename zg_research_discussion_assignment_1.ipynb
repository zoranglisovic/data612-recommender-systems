## Part 1 — How Netflix's Recommender System Works

I'm choosing Netflix as my commercial recommender — a platform I have a personal connection to, having worked in the VFX industry on several Netflix original productions. From that experience, I saw firsthand how recommendation performance isn't just a user experience feature — it directly influences which content gets greenlit, renewed, or cancelled.

### How it works

Netflix uses a **hybrid recommender system** that combines multiple techniques simultaneously:

**Collaborative Filtering** is the backbone. Netflix looks at what users with similar taste profiles have watched and rated, and uses that overlap to predict what you might enjoy. This is essentially what the Netflix Prize was trying to optimize — the winning algorithms were all collaborative filtering variants based on matrix factorization (SVD).

**Content-Based Filtering** supplements this by analyzing metadata about the content itself — genre, director, cast, themes, even visual style. If you watch a lot of crime dramas with a specific director, Netflix can surface other titles with similar attributes even if no other user has connected them.

**Contextual signals** add another layer — what device you're on, what time of day it is, how long your session has been. Netflix has published research showing that recommendations on a TV at 9pm should look different from recommendations on a phone during a commute.

**Artwork personalization** is something less obvious but very real. Netflix doesn't just recommend a title — it shows you a different thumbnail image for that title based on your watch history. A user who watches a lot of action films might see an explosion-heavy thumbnail for a romantic comedy, while another user sees the lead actors. This is a form of recommendation layered on top of recommendation.

### Does it deliver a good experience?

Mostly yes — Netflix's recommender is genuinely impressive at keeping users engaged. From my VFX work perspective though, I noticed something interesting: the system can create a feedback loop where content that performs well in recommendations gets more budget, which produces better content, which performs better again. This can make it harder for genuinely novel or risky projects to break through unless a showrunner or executive champions them past the algorithm.

The biggest weakness I've observed as a user is the **cold start problem** — when a new show launches, there's limited data to personalize recommendations for it, so Netflix relies heavily on demographic assumptions and genre placement. This can cause quality content to underperform simply because it doesn't fit a clean category.

## Part 2 — Attacks on Recommender Systems

### The Article

Travis M. Andrews' 2017 Washington Post article describes how IMDb users coordinated to destroy the rating of *The Promise* — a film starring Christian Bale and Oscar Isaac, directed by Terry George (*Hotel Rwanda*) — before it even opened in theaters.

The film is set against the backdrop of the Armenian Genocide, depicting the mass killing of an estimated 1.5 million Armenians by the Ottoman Empire beginning in 1915. When word of the film's upcoming release spread to a Turkish message board, users organized to flood IMDb with 1-star reviews to tank the movie before anyone could see it. One user posted: *"Guys, Hollywood is filming a big movie about the so-called Armenian genocide and the trailer has already been watched 700k times. We need to do something urgently."*

By the time the film premiered at the Toronto Film Festival — to a standing ovation — it already had nearly 120,000 IMDb ratings, almost 62,000 of them one-star. At its lowest point it had a 1.8 rating, placing it among the 10 worst-rated films on the entire site. Producer Mike Medavoy summed it up: *"All I know is that we were in about a 900-seat house with a real ovation at the end, and then you see almost 100,000 people who claim the movie isn't any good."*

This is a textbook case of **review bombing** — coordinated, politically motivated manipulation of a rating system by people who had not seen the content, with the goal of suppressing it.

### A Similar Real-World Example

The article itself mentions several other examples — Amy Schumer's Netflix special was targeted by Trump supporters on Reddit, and the female-led *Ghostbusters* reboot was flooded with negative reviews from men opposed to the casting, throwing the averages "completely out of whack."

Another well-known case is the **GamerGate campaigns** of 2014–2016, where coordinated groups targeted video games and public figures on Metacritic, Steam, and IMDb — with thousands of extreme scores appearing within hours, driven entirely by ideology rather than the quality of the work.

What all these cases have in common is that the attack isn't really about the content — it's about using a rating system as a political or social weapon.

### How I Would Design a System to Prevent This

There is probably no perfect solution, but a few common-sense changes could make a big difference:

**1. Require proof of viewing** — Only allow ratings from users who have actually watched the content. Netflix already does this since they track everything you watch. IMDb or Metacritic could require a verified purchase or a minimum watch time before you're allowed to leave a rating. This alone would eliminate most review bombing since the whole point is that people are rating things they haven't seen.

**2. Flag unusual rating spikes** — If a movie gets hundreds of 1-star ratings in a few hours, especially before wide release, something is clearly off. The system could automatically flag this and hold those ratings for review, or simply not display the score until it stabilizes.

**3. Treat new accounts with suspicion** — A brand new account that immediately submits an extreme rating during a spike should count for less than a long-standing account with a normal rating history. If someone creates an account just to bomb a movie, their vote shouldn't carry the same weight as a regular user.

**4. Separate critic and audience scores** — Rotten Tomatoes already does this well. Keeping professional reviews separate from audience scores gives people two independent signals to look at. A coordinated audience attack won't affect the critic score, so at least one number remains trustworthy.

The core problem is that any open rating system can be gamed. The goal is just to make it hard enough that it's not worth the effort.
