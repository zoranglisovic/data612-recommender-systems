**Topic:** Mitigating the Harm of Recommender Systems

The three readings agree on the surface-level diagnosis — recommendation engines optimize for engagement, and engagement-optimized systems tend to surface increasingly extreme content — but I found the Krishnan et al. paper the most useful for thinking about why interventions are so hard to verify.

Their core finding is that displaying an aggregate rating before a user submits their own shifts that rating toward the displayed median. In their study, ratings moved an average of 19.3% closer to the median in users who saw the aggregate first. That's social influence bias — and it gets recorded in the training data as ground truth. The model then trains on those shifted ratings, produces recommendations calibrated to a median that was artificially inflated, and the next round of user interactions are generated in response to those recommendations. The bias doesn't stay in one round; it compounds.

This creates a problem that I think is easy to overlook: if historical ratings are already biased by social influence, then offline evaluation metrics like RMSE and precision are being measured against corrupted ground truth. You could add a diversity penalty or de-prioritize extreme content, run an offline evaluation, and see improvement — but you'd be measuring improvement against a baseline that was already distorted. The only way to really verify whether a harm-mitigation change works is an online A/B test, but running one long enough to detect slow radicalization effects means deliberately exposing a control group to the unmodified system for the duration.

Krishnan's correction model — estimating what ratings would have been without the social influence — is a reasonable first step toward cleaning up historical data before training or evaluation. DiResta's argument that platforms need to treat recommendation as an editorial act with accountability, rather than a neutral optimization, points at the same underlying issue from a governance angle: if you don't acknowledge that the feedback loop is corrupting your signal, you won't invest in fixing it.

Do you think platforms would be willing to collect the pre/post rating triplets Krishnan's method requires, given that doing so makes their social influence problem visible and measurable?
